Culminated list of news source, looking for all href in each page which have date in it.
Current date and previous date news will bbe considered, with high weightage to current day

```
news_source = ['https://economictimes.indiatimes.com/industry',
               'https://www.moneycontrol.com/news/business/markets/',
               'https://www.moneycontrol.com/news/business/stocks/',
               'https://www.moneycontrol.com/news/business/companies/',
               'https://in.investing.com/news/',
               ]
```

```
news = []
articles = []
quote=[]
header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'}

for url in news_source:
    res = requests.get(url, headers= header)
    soup = bs4.BeautifulSoup(res.text,'html.parser')
    body = soup.body

    if 'economictimes' in url:
        base = 'https://economictimes.indiatimes.com'
    
    elif 'investing.com' in url:
        base = 'https://in.investing.com'
    
    else:
        base = ''
    
    if 'investing.com' not in url:
        for links in body.find_all('a'):
            if links.get('href'):
                
                # Loop for economictimes
                if 'http' not in links.get('href') and '/industry/' in links.get('href') and base:
                    
                    #print(links)
                    news.append(base+links.get('href'))
                
                #Loop for moneycontrol
                elif 'moneycontrol' in links.get('href') and url in links.get('href'):
                    #print(links)
                    news.append(links.get('href'))       
                
    else:
        #Loop for investing.com
        links = body.find_all("a", class_="link")
        for link in links:
            if link.get('href'):
                #print(link.get('href'))
                news.append(base+ link.get('href'))

for l in news:        
    news_content = newspaper(l)
    #quote.append(json.dumps(news.get_dict))
    content = news_content.article
    articles.append(content)
    print(l)
    #print(content)
```

```
articles = list(set(articles))
articles.remove('Find this comment offensive? Choose your reason below and click on the Report button. This will alert our moderators to take action Name Reason for reporting: Foul language Slanderous Inciting hatred against a certain community')
print(articles)
```

```
sid.lexicon
```

```
score={'positive':0,
       'negative':0,
       'neutral':0,
       'overall':0,
       }

positive, negative = 0,0

f = open('News score.txt', 'a')

for i in articles:
    print(i)
    intensity = sid.polarity_scores(i)
    print(intensity)
    score['positive'] += intensity['pos']
    score['negative'] += intensity['neg']
    score['neutral'] += intensity['neu']
    score['overall'] += intensity['compound']
    if intensity['compound'] >= 0.1:
        positive+=1
    else:
        negative+=1
    print()
    f.write(i+'\n')
    f.write(str(intensity)+'\n')
    f.write('\n')

score['positive'] /= len(articles)
score['negative'] /= len(articles)
score['neutral'] /= len(articles)
score['overall'] /= len(articles)

f.close()
print(score)
print(f'Total positive news: {positive}; Total negative news: {negative}')
```
